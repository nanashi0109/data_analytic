```python
import numpy as np
from matplotlib import pyplot as plt
import csv
from collections import Counter
```

```python
def get_data(path):
    with open(path, "r", newline="", encoding = "utf-8") as file:
        reader = csv.reader(file)

        collumns = [[], [], []]
        for row in reader:
            collumns[0].append(row[0])
            collumns[1].append(row[1])
            collumns[2].append(row[2])
        collumns[0].pop(0)
        collumns[1].pop(0)
        collumns[2].pop(0)

        return collumns
        
data = get_data("delivery_times.csv")
```

```python
order_number = np.array(data[0], dtype="int8")
time_a = np.array(data[1], dtype="int8")
time_b = np.array(data[2], dtype="int8")
```

### **Среднее время доставки службой А**

```python
time_a.mean()
```

### **Среднее время доставки службой Б**

```python
time_b.mean()
```

### **Медиана времени доставки службой А**

```python
np.median(time_a)
```

### **Медиана времени доставки службой Б**

```python
np.median(time_b)
```

### **Стандартное отклонение времени доставки службой А**

```python
np.sqrt(time_a.var())
```

### **Стандартное отклонение времени доставки службой А**

```python
np.sqrt(time_b.var())
```

### **Графики**

```python
plt.bar(order_number, time_a)

plt.title("Время доставки заказов службы А")
plt.xlabel("Заказ")
plt.ylabel("Время")

plt.show()
```

```python
plt.bar(order_number, time_b)

plt.title("Время доставки заказов службы Б")
plt.xlabel("Заказ")
plt.ylabel("Время")

plt.show()
```

```python
plt.plot(order_number, time_a, color="lightblue", marker="o")

plt.plot(order_number, time_b, color="red", marker="o")

plt.title("График времени доставки службы А и Б")
plt.xlabel("Заказ")
plt.ylabel("Время")
```
