```python
import numpy as np
from matplotlib import pyplot as plt
import csv
from collections import Counter
```

```python
def get_data(path):
    with open(path, "r", newline="", encoding = "utf-8") as file:
        reader = csv.reader(file)

        collumns = [[], [], []]
        for row in reader:
            collumns[0].append(row[0])
            collumns[1].append(row[1])
            collumns[2].append(row[2])
        collumns[0].pop(0)
        collumns[1].pop(0)
        collumns[2].pop(0)

        return collumns
        
data = get_data("grades.csv")
```

```python
names = np.array(data[0])
math_grades = np.array(data[1], dtype="int8")
lit_grades = np.array(data[2], dtype="int8")
```

### **Средний балл по математике**

```python
math_grades.mean()
```

### **Средний балл по литературе**

```python
lit_grades.mean()
```

### **Медиана оценок по математике**

```python
np.median(math_grades)
```

### **Медиана оценок по литературе**

```python
np.median(lit_grades)
```

### **Стандартное отклонение оценок по математике**

```python
np.sqrt(math_grades.var())
```

### **Стандартное отклонение оценок по литературе**

```python
np.sqrt(lit_grades.var())
```

### **Графики**

```python
plt.bar(names, math_grades)

plt.title("Оценки по математике")
plt.ylabel("Оценки")
plt.xlabel("Студент")

plt.show
```

```python
plt.bar(names, lit_grades)

plt.title("Оценки по литературе")
plt.ylabel("Оценки")
plt.xlabel("Студент")

plt.show
```
